```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"../data/processed_data/bank_of_canada_cleaned.csv")
```

```python
numeric_cols = df.select_dtypes(include='number').columns
categorical_cols = df.select_dtypes(include='object').columns
```

```python
months = ['May-25', 'Jun-25', 'Jul-25', 'Aug-25', 'Sep-25']

for i in range(1, len(months)):
    df[f'{months[i]}_MoM_change'] = df[months[i]] - df[months[i-1]]
```

```python
for i in range(1, len(months)):
    df[f'{months[i]}_MoM_pct_change'] = (
        (df[months[i]] - df[months[i-1]]) / df[months[i-1]]
    ) * 100
```

```python
df['avg_credit'] = df[months].mean(axis=1)
df['max_credit'] = df[months].max(axis=1)
df['min_credit'] = df[months].min(axis=1)
df['credit_range'] = df['max_credit'] - df['min_credit']
```

```python
df['credit_trend'] = df[months[-1]] - df[months[0]]
```

```python
df['credit_std_dev'] = df[months].std(axis=1)
```

```python
#If multiple credit categories
df['household_to_total_credit'] = (
    df['Household Credit'] / df['Total Credit']
)
```

```python
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
```
